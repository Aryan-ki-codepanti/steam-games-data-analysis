```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

data_file_path = "../data/final_0_80509.csv"
```

```python
df = pd.read_csv(data_file_path,parse_dates=True, low_memory=False)
df.info()
```

```python
df = df[["AppID","Title","Interface_Languages","Audio_Languages","Subtitle_Languages","Current_Players"]]
df.info()
df.head(10)
```

```python
df = df.dropna()
df.info()
df.head(10)
```

```python
def count_languages_length(languages_str):
    languages_list = languages_str.split(',')
    return len(languages_list)

df['Audio_Languages_Length'] = df['Audio_Languages'].apply(count_languages_length)
df['Interface_Languages_Length'] = df['Interface_Languages'].apply(count_languages_length)
df['Subtitle_Languages_Length'] = df['Subtitle_Languages'].apply(count_languages_length)

df['Total_Languages_Count'] = df['Interface_Languages_Length'] + df['Audio_Languages_Length'] + df['Subtitle_Languages_Length']
non_zero_players = df['Current_Players']!=0
correlation = df['Total_Languages_Count'].corr(non_zero_players)
print(correlation)
print(non_zero_players.head(10))
```
