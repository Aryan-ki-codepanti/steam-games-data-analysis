## Determine the market share of the top developers and publishers on STEAM based on the number of games. Visualize this information in a meaningful way.

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

data_file_path = "../data/final_0_80509.csv"
```

```python
df = pd.read_csv(data_file_path,parse_dates=True, low_memory=False)
df.info()
```

```python
df = df[["AppID","Title","Developers","Publishers"]]
df.info()
df.head(10)
```

```python
df = df.dropna()
df.info()
df.head(10)
```

```python
dev_df = df['Developers'].value_counts()
print(dev_df.head(20))

pub_df = df['Publishers'].value_counts()
print(pub_df.head(20))
```

```python
skip = None
top_dev = dev_df.head(50)

plt.figure(figsize=(10, 10))
plt.bar(top_dev.index, top_dev.values, label='Top Developers',color = 'red')
plt.xticks(rotation=90)
plt.xlabel('Developers',fontsize = 15)
plt.ylabel('Number of Games',fontsize = 15)
plt.title('Market Share of Top 50 Developers')
plt.legend(loc = 1, fontsize = 15)
plt.show()
```

```python
skip = None
top_pub = pub_df.head(50)

plt.figure(figsize=(10, 10))
plt.bar(top_pub.index, top_pub.values, label='Top Publishers',color = 'green')
plt.xticks(rotation=90)
plt.xlabel('Publishers',fontsize = 15)
plt.ylabel('Number of Games',fontsize = 15)
plt.title('Market Share of Top 50 Publishers')
plt.legend(loc = 1, fontsize = 15)
plt.show()
```

### Inferences

1. 'Big Fish Games' emerges as the top publisher in the market.
2. 'Choice of Games' emerges as the top developer in the market.
3. Only 7 developers have a market value of 100+ games.
4. Only 6 producers have a market value of 200+ games.
