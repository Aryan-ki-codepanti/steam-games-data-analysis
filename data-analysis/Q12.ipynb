Distribution of games based on the WIN_FLAG, MAC_FLAG, LINUX_FLAG (i.e. how many games are compatible with all three, then how many are compatible with two of them, then how many are compatible with only one, covering all the combinations of the three)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/final_0_80509.csv")
```

```python
win_games = df[df['Win_Flag'] == 1].shape[0]
mac_games = df[df['Mac_Flag'] == 1].shape[0]
linux_games = df[df['Linux_Flag'] == 1].shape[0]
platforms = ['WIN', 'MAC', 'LINUX']
counts = [win_games, mac_games, linux_games]

bars = plt.bar(platforms, counts, color='skyblue', width=0.3)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')
plt.xlabel('Platforms')
plt.ylabel('Number of Games')
plt.title('Number of Games for Different Platforms')
plt.show()
```

As we can see, number of games made for Windows(78952) is exponentionally higher than that for any other platform, MAC(16268) and Linux(11368)

```python
only_win = df[(df['Win_Flag'] == 1) & (df['Mac_Flag'] == 0) & (df['Linux_Flag'] == 0)].shape[0]
only_mac = df[(df['Win_Flag'] == 0) & (df['Mac_Flag'] == 1) & (df['Linux_Flag'] == 0)].shape[0]
only_linux = df[(df['Win_Flag'] == 0) & (df['Mac_Flag'] == 0) & (df['Linux_Flag'] == 1)].shape[0]

both_win_mac = df[(df['Win_Flag'] == 1) & (df['Mac_Flag'] == 1) & (df['Linux_Flag'] == 0)].shape[0]
both_win_linux = df[(df['Win_Flag'] == 1) & (df['Mac_Flag'] == 0) & (df['Linux_Flag'] == 1)].shape[0]
both_mac_linux = df[(df['Win_Flag'] == 0) & (df['Mac_Flag'] == 1) & (df['Linux_Flag'] == 1)].shape[0]

all_platforms = df[(df['Win_Flag'] == 1) & (df['Mac_Flag'] == 1) & (df['Linux_Flag'] == 1)].shape[0]

platforms = ['Only WIN', 'Only MAC', 'Only LINUX', 'WIN and MAC', 'WIN and LINUX', 'MAC and LINUX', 'All Platforms']
counts = [only_win, only_mac, only_linux, both_win_mac, both_win_linux, both_mac_linux, all_platforms]
j = 0
for i in platforms:
    print(i,' -> ',counts[j])
    j+=1
```

```python
bars = plt.bar(platforms, counts, color='skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')
plt.xlabel('Platforms and Combinations')
plt.ylabel('Number of Games')
plt.title('Number of Games for Different Platforms and Combinations')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

Individual Platforms:
- The majority of games are compatible only with Windows (WIN), with a substantial count of 59,896.
- Mac (MAC) has a very small number of exclusive games, with only 20 titles.
- Linux (LINUX) has even fewer exclusive games, with only 5 titles.

Two Platform Combinations:
- There is a notable number of games that are compatible with both Windows and Mac (WIN and MAC), totaling 7,334.
- Games compatible with both Windows and Linux (WIN and LINUX) are also substantial, with 2,449 titles.
- There is only one game that is compatible with both Mac and Linux (MAC and LINUX).

All Three Platforms:
- A significant number of games, 8,913 in total, are compatible with all three platforms (All Platforms).

Conclusion:
The dataset reflects a clear dominance of games designed exclusively for Windows, with relatively limited support for Mac and Linux individually. However, there is a noteworthy number of games that cater to users of both Windows and Mac or Windows and Linux. Compatibility with all three platforms is present in a substantial number of titles, suggesting a growing trend in cross-platform game development. Developers and users interested in diverse platform support should find this information valuable for making informed decisions.
