## Loading Modules and functions

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

## Loading Data

```python
df = pd.read_csv("../processed.csv")
df
```

## Splitting Data 33% test, 66% train

```python
y = df["target"]
X = df.drop(labels=["target"],axis=1)


X_train, X_test, y_train, y_test = train_test_split(
         X, y, test_size=0.33, random_state=100,shuffle=True)
```

## Fitting model

```python
bayes = GaussianNB()
bayes.fit(X_train, y_train)
```

## Predicting Likelihood of test cases

```python
bayes.predict_proba(X_test)
```

## Model evaluation

```python
preds = bayes.predict(X_test)
print(f"Accuracy : {accuracy_score(y_test, preds)}")
```

## Important metrics

```python
print(classification_report(y_test, preds))
```

## Confusion Matrix Display

```python
confusion_matrix(y_test,preds)
```

## 1st trial Conclusion
1. Bayes Classifier gave *worst* performance of 18% accuracy
